```python
# Produce a figure showing covariate shift
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
np.random.seed(1)
# Generate data
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15, 6))

# Leave space between rows
fig.subplots_adjust(hspace=0.5)

train_sample_distribution = np.random.normal(-2, 2, 1000)
test_sample_distribution = np.random.normal(2, 2, 1000)

# Plot a random function y = something of x with some noise
x = np.linspace(-10, 10, 100)
y = 0.05*x**3+50

ax1.plot(x, y)
ax1.set_xlabel('Feature value')
ax1.set_ylabel('Target value')
ax1.set_title('Covariate Shift Example')
ax1.legend(['Target Function'])

# Plot train and test sample distributions
ax1.hist(train_sample_distribution, bins=100)
ax1.hist(test_sample_distribution, bins=100, alpha=0.5)
ax1.set_xlabel('Feature value')
ax1.set_ylabel('Frequency')
ax1.legend(['Target Function', 'Train Sample', 'Test Sample'])



train_sample_distribution = np.random.normal(0, 3, 1000)
test_sample_distribution = np.random.normal(0, 3, 1000)
ax2.plot(x, y)
yy = -4 * x + 50# + np.random.normal(0, 5, 100)
ax2.plot(x, yy, color='red')
# Add small arrows to indicate shift from y to yy for every point in a 1 point interval
for i in range(0, len(x), 5):
    ax2.annotate('', xy=(x[i], yy[i]), xytext=(x[i], y[i]),
                 arrowprops=dict(facecolor='black', shrink=0.05))

ax2.set_xlabel('Feature value')
ax2.set_ylabel('Target value')
ax2.set_title('Concept Shift Example')
ax2.legend(['Target Function'])

# Plot train and test sample distributions
ax2.hist(train_sample_distribution, bins=100)
ax2.hist(test_sample_distribution, bins=100, alpha=0.5)
ax2.set_xlabel('Feature value')
ax2.set_ylabel('Frequency')
ax2.legend(['Target function','Shifted Function', 'Train Sample', 'Test Sample'])

fig.subplots_adjust(hspace=0.5)
ax1.set_xlim(-10, 10)
ax1.set_ylim(ax1.get_ylim())

ax2.set_xlim(-10, 10)
ax2.set_ylim(ax2.get_ylim())
plt.savefig('covariate_shift_data.png')
plt.show()
```
